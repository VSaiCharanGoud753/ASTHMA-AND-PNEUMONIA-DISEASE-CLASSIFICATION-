```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\vithe\OneDrive\Desktop\Mini Project\DemographicData.csv")
data
```

```python
x = data['Country Name'].head()
x
```

```python
y = data['Birth rate'].head()
y
```

```python
k = data['Internet users'].head()
k
```

```python
i = data['Income Group'].head()
i
```

```python
plt.scatter(x,y,marker = '*',color = 'blue',label = 'data')
plt.xlabel('Country name')
plt.ylabel('Birth Rate')
plt.title('Scatter plot of Birth rate of each Country')
plt.grid()
plt.legend()
plt.show()
```

```python
#Line Plot
```

```python
plt.plot(x,y, marker = 'o',color = 'red', label = 'data')
plt.xlabel('Country name')
plt.ylabel('Birth Rate')
plt.title('Line plot of Birth rate of each Country')
#plt.grid(True)
plt.legend()
plt.show()
```

```python
#Bar plot
```

```python
plt.bar(x,y, color = 'green', label = 'data')
plt.xlabel('Country name')
plt.ylabel('Birth Rate')
plt.title('Bar plot of Birth rate of each Country')
plt.legend()
plt.show()
```

```python
#Box Plot
```

```python
data = np.random.randn(5,10)
plt.boxplot(data)
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot')
#plt.grid(True)
plt.show()
```

```python
#Heat Map
```

```python
plt.imshow(data, cmap = 'rainbow')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

```python
#Scatter Bar Plot
```

```python
categories = ['A','B','C','D']
values1 =[20,12,13,42]
values2 = [17,16,20,20]
plt.bar(categories, values1, color = 'Blue', label = 'data 1')
plt.bar(categories, values2, color = 'orange', bottom = values1, label = 'data 2')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Stacked Bar plot')
plt.legend()
plt.show()
```

```python
#Violin Plot
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data, showmedians=True)
plt.xlabel('Distribution')
plt.ylabel('Value')
plt.title('Violin Plot')
#plt.xticks(np.arange(1, len(data) + 1), ['Data 1', 'Data 2', 'Data 3'])
plt.legend(['Median'])
plt.show()
```
